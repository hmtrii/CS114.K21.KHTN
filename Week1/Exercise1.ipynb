#**PROGRAMING**

```
def maxSubArraySum(a): 
  size = len(a)
  maxGlobal = - 1
  maxCurrent = 0
  start = 0
  end = 0
  s = 0
  
  for i in range(0,size):   
    a[i] = int(a[i])
    maxCurrent += a[i] 

    if maxGlobal < maxCurrent:
      maxGlobal = maxCurrent 
      start = s 
      end = i 

    if maxCurrent < 0:
      maxCurrent = 0
      s = i+1
  return (start+1, end+1, maxGlobal)

#Start input
n = int(input())

a = []
a = input().split(" ")
#Output
(p,q,r) = maxSubArraySum(a)
print(p,q,r)
```
